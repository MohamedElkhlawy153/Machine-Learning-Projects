### Data Preprocessing

```python
# encoding data by use mapping
map_data={
    'Graduate':1,
    'Dropout':0,
    'Enrolled':2
}
data['Target']=data['Target'].map(map_data)
```

```python
# count value in target class
data['Target'].value_counts()
```

```python
# select only Graduate and Dropout from data as numper of Enrolled very low
data=data[data['Target']!=2]
```

```python
# split data into x and y
x=data.drop(['Target'],axis=1)
y=data['Target']
```

```python
x.head()
```

```python
y.head()
```

```python
# scale data by use minmaxscaler
cols=x.columns

from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
x=Scaler.fit_transform(x)
```

```python
x=pd.DataFrame(x,columns=cols)
```

```python
# apply chi2,kbest to choose best feature
from sklearn.feature_selection import SelectKBest,chi2

Select_KB=SelectKBest(chi2,k=15)
x_new=Select_KB.fit_transform(x,y)
```

```python
x_new.shape
```

```python
# show 15 best feature
print(x.columns[Select_KB.get_support()].to_list())
```

```python
x=x_new
```

```python
# show corrlation on all 15 best feature and y target
data=np.column_stack([x,y])
data=pd.DataFrame(data)
plt.figure(figsize=(20,20))
sns.heatmap(data.corr(),annot=True,fmt='.3f')
plt.show()
```

```python
# split data int0 train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_test.shape)
print(y_test.shape)
```

### Build Model

### 1-Decision Tree

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
DT=DecisionTreeClassifier(random_state=42)

params={'splitter': ["best", "random"],
          'max_depth': np.arange(1,10,1),
          'min_samples_leaf': np.arange(1,5,1),
          'min_samples_split': np.arange(2,5,1)
         }
grid=GridSearchCV(estimator=DT,param_grid=params,cv=5,scoring='accuracy',n_jobs=-1)
grid.fit(x_train,y_train)
best_p=grid.best_params_
print(best_p)
grid.best_score_
```

```python
# use best parameter and calculate preformance
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix

DT_model=DecisionTreeClassifier(**best_p,random_state=42)
DT_model.fit(x_train,y_train)
y_pred_test=DT_model.predict(x_test)
y_pred_train=DT_model.predict(x_train)

print('test accuracy =',accuracy_score(y_test,y_pred_test))
print('train accuracy =',accuracy_score(y_train,y_pred_train))
```

```python
print('test classification_report =',classification_report(y_test,y_pred_test))
print(40*'--')
print('train classification_report =',classification_report(y_train,y_pred_train))
```

```python
# plot confusion_matrix
cm=confusion_matrix(y_test,y_pred_test)
plt.figure(figsize=(5,5))
sns.heatmap(cm,annot=True,fmt='.0f',cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,y_pred_test))
```

### 6-Stacking Classifier

```python
# use stacking
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import SGDClassifier

svc = SVC(probability=True)
lr = LogisticRegression(C=.1)
rf = RandomForestClassifier(**best_pp)
dt = DecisionTreeClassifier(**best_p)
xgb = XGBClassifier(**best_ppp)
mlp = MLPClassifier(**best_pppp)
bagging = BaggingClassifier()
adaboost = AdaBoostClassifier()
sgd=SGDClassifier(alpha= 0.01)
# define the base models
level0 = list()
level0.append(('LR', lr))
level0.append(('RF',rf))
level0.append(('DT',dt))
level0.append(('SGD',sgd))
level0.append(('XGB',xgb))
level0.append(('MLP',mlp))
level0.append(('Adaboost',adaboost))
level0.append(('bagging',bagging))
# define meta learner model
level1= SVC(C= 0.7999999999999999, gamma= 0.04, kernel= 'rbf')
# define the stacking ensemble
stacking = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)
# fit the model on all available data
stacking.fit(x_train, y_train)
y_pred_test_stack = stacking.predict(x_test)
y_pred_train_stack = stacking.predict(x_train)
```

```python
print('Test Accuracy =', accuracy_score(y_test, y_pred_test_stack))
print('Train Accuracy =', accuracy_score(y_train, y_pred_train_stack))
```

```python
print('test classification_report =',classification_report(y_test,y_pred_test_stack))
print(40*'--')
print('train classification_report =',classification_report(y_train,y_pred_train_stack))
```

```python
# plot confusion_matrix
cm=confusion_matrix(y_test,y_pred_test_stack)
plt.figure(figsize=(5,5))
sns.heatmap(cm,annot=True,fmt='.0f',cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,y_pred_test_stack))
```
