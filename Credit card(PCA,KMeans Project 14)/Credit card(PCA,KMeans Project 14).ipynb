### 3-Data Preprocessing

```python
# drop non important columns
data.drop(['CUST_ID'],axis=1,inplace=True)
```

```python
data.head()
```

```python
# handle missing values
from sklearn.impute import SimpleImputer
Imputer=SimpleImputer(strategy='mean',missing_values=np.nan)
data[['CREDIT_LIMIT','MINIMUM_PAYMENTS']]=Imputer.fit_transform(data[['CREDIT_LIMIT','MINIMUM_PAYMENTS']])
```

```python
# checking missing values
data.isnull().sum()
```

```python
# scalling data 
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
data_scale=Scaler.fit_transform(data)
```

```python
data_scale
```

### 4-Apply Model

```python
# apply PCA
from sklearn.decomposition import PCA
pca_model=PCA()
pca_model.fit(data_scale)
```

```python
# show number of important feature
plt.plot(np.cumsum(pca_model.explained_variance_ratio_))
plt.xlim(0,8)
plt.xlabel('Number of Components')
plt.ylabel('variance')
plt.show()
```

```python
# pca--> n features = 8
pca_model=PCA(n_components=8)
pca_data=pca_model.fit_transform(data_scale)
```

```python
pca_data
```

```python
pca_data.shape
```

```python
from yellowbrick.cluster import KElbowVisualizer,silhouette_visualizer
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
# elbow method
elbow_method=KElbowVisualizer(KMeans(),k=10)
elbow_method.fit(pca_data)
elbow_method.show()
```

```python
# by use another way
scores=[]
silhouette_s=[]
k_range=range(2,10)
for k in k_range:
    km=KMeans(n_clusters=k,random_state=42)
    y_pred=km.fit_predict(pca_data)
    scores.append(km.inertia_)
    silhouette_s.append(silhouette_score(pca_data,y_pred))
    print('k = ',k ,'  ',km.inertia_)
    print('silhouette_score : %0.4f'%silhouette_score(pca_data,y_pred))
```

```python
# plot Elbow method
plt.plot(k_range,scores,marker='o')
plt.title('Elbow Method')
plt.xlabel('K Number')
plt.ylabel('k_means Distance')
plt.show()
```

```python
# apply model on oneway k=3
km_model=KMeans(3,random_state=42)
visualizer=silhouette_visualizer(km_model,pca_data,color='yellowbrick')
visualizer.fit(pca_data)
visualizer.show()
```

```python
# best k = 3
kmeans_model=KMeans(n_clusters=3,random_state=42)
y_pred_km=kmeans_model.fit_predict(pca_data)
```
